## Origin Airports

```python
#Similar to edX problem notebook 11:
'''Aggregation. Observe that an (origin, destination) pair may appear many times. 
That's because the dataset includes a row for every direct flight that occurred historically 
and there may have been multiple such flights on a given day.
However, for the purpose of this analysis, let's simplify the problem by collapsing all 
historical segments  iâjiâj  into a single segment. Let's also do so in a way that preserves the 
number of times the segment occurred (i.e., the number of rows containing the segment).'''

flights_cols_subset = flights[['FL_DATE', 'ORIGIN_AIRPORT_ID','ORIGIN', 'DEST_AIRPORT_ID', 'DEST']]
segment_groups = flights_cols_subset.groupby(['ORIGIN_AIRPORT_ID', 'ORIGIN', 'DEST_AIRPORT_ID', 'DEST'], as_index=False)
segments = segment_groups.count()
segments.rename(columns={'FL_DATE': 'FL_COUNT'}, inplace=True)
segments.head()
```

```python
assert (segments['FL_COUNT'] > 0).all() # verify that the counts are all at least 1
```

```python
#What are our actual origin locations
origins = segments[['ORIGIN_AIRPORT_ID', 'ORIGIN', 'FL_COUNT']].groupby(['ORIGIN_AIRPORT_ID','ORIGIN'], as_index=False).sum()
origins.rename(columns={'FL_COUNT': 'ORIGIN_COUNT'}, inplace=True)
print("Number of actual origins:", len(origins))
origins.head()
```

To get an idea of importance, let's rank the airports by the total number of outgoing flight segments.
i.e flight segments that originate at the airport

```python
#Origins Ranking by Outgoing flights
origins_OUT_ranking = origins.sort_values(by = ['ORIGIN_COUNT'], ascending = False)
origins_OUT_ranking = pd.merge(origins_OUT_ranking, airports, how = 'left',\
                         left_on = ['ORIGIN'], right_on = ['IATA_CODE']).drop(['IATA_CODE'], axis = 1)

# Prints the origin airports, according to your calculation:
origins_OUT_ranking
```

```python
# DEFINE COORDINATE RANGE IN WEB MERCATOR COORDINATE SYSTEM (EPSG:3857)
#This is US range
x_range,y_range=([-15187814,-6458032], [2505715,6567666])
```

```python
#Plot the airport position on the map
#But hold on! Latitude and longitude is in geographic coordinate system (EPSG:4326). 
#On the other hand, the map is in web mercator coordinate system. The coordinate system is not match 
#to each other, so we can not plot the aircraft position directly on the map. 
#The aircraft position must be converted to web mercator coordinate system. 
#Therefore we need to make a function to convert the latitude and longitude to mercator y and x. 
#Later in the code I called this function  as wgs84_to_web_mercator.
```

```python
#To plot points... 
#Step 1:  convert the pandas df into bokeh column data source
#Step 2: Add the points using bokeh figure circle method

#FUNCTION TO CONVERT GCS WGS84 TO WEB MERCATOR
#https://www.geodose.com/2019/01/realtime-flight-tracking-pandas-bokeh-python.html
def wgs84_to_web_mercator(df, lon="LONGITUDE", lat="LATITUDE"):
    k = 6378137
    df["x"] = df[lon] * (k * np.pi/180.0)
    df["y"] = np.log(np.tan((90 + df[lat]) * np.pi/360.0)) * k
    return df
wgs84_to_web_mercator(origins_OUT_ranking)
```

```python
#Notice that there are two new columns, x, y that represent the WEB MERCATOR value
#Now, add in a scaled count column for the "bubble" effect based on importance
scale = 700
origins_OUT_ranking['ORIGIN_COUNT_SCALED'] = origins_OUT_ranking['ORIGIN_COUNT']/scale
origins_OUT_ranking.head(5)
```

```python
#Visualize the Airport Importance by Number of Outgoing Flights
#Resource:  https://www.geodose.com/2019/01/realtime-flight-tracking-pandas-bokeh-python.html

#SET COLOR PALETTE
color_mapper = linear_cmap(field_name='ORIGIN_COUNT', palette = cividis(6) ,low=min(origins_OUT_ranking['ORIGIN_COUNT']),\
                                    high=max(origins_OUT_ranking['ORIGIN_COUNT']))

#BOKEH COLUMN DATA SOURCE
flight_source=ColumnDataSource(origins_OUT_ranking)

#SETUP FIGURE
p=figure(x_range=x_range,y_range=y_range,x_axis_type='mercator',y_axis_type='mercator',
  sizing_mode='scale_width',plot_height=300,title = "Airports by Number of Outgoing Flights")
p.add_tile(tile_provider)

#Add Airports
p.circle('x','y',line_color = color_mapper, color = color_mapper, source=flight_source,\
         size= 'ORIGIN_COUNT_SCALED',fill_alpha=1,line_width=0.5)

#SETUP HOVER TOOL
my_hover = HoverTool()
my_hover.tooltips = [('Origin', '@ORIGIN'),('Outgoing_Flights', '@ORIGIN_COUNT')]
p.add_tools(my_hover)

#Setup Color Bar
color_bar = ColorBar(color_mapper = color_mapper['transform'], width = 25, location = (0,0))
p.add_layout(color_bar, 'right')
               
#SHOW FIGURE INLINE
show(p)

#OPEN FIGURE IN BROWSER
p
```

# Construct the state-trasistion matrix


Next, we construct the probability state-iteration matrix for the airport network.  
We want the conditional probability (Pij) that a random flyer departs from airport i and arrives at airport j, 
given that he/she is currently at airport i.

```python
n_airports = airports.index.max() + 1
print("Note: There are", n_airports, "airports in our airport set.")
```

```python
#add a new column to the segments dataset that is the integer index of the airport
# Do this for both the ORIGIN and DEST airports
origin_indices = airports[['IATA_CODE']].rename(columns={'IATA_CODE' : 'ORIGIN'})
origin_indices['ORIGIN_INDEX'] = airports.index

dest_indices = airports[['IATA_CODE']].rename(columns={'IATA_CODE' : 'DEST'})
dest_indices['DEST_INDEX'] = airports.index

#troubleshooting in case I run code multipel times... 
if 'ORIGIN_INDEX' in segments.columns:
    del segments['ORIGIN_INDEX']
    
#Merge index to the segments
segments = segments.merge(origin_indices, on = 'ORIGIN', how = 'left')
segments = segments.merge(dest_indices, on = 'DEST', how = 'left')

segments.head()
```

Now,  compute the edge weights for the flight segments  (or the probabiliyt Pij)
For each origin, i, let di be the number of outgoing edges, or 'outdegree'.  This is NOT the same as the total number of (historical) outbound segments, rather,dj is just the number of airports directly reachable from i.   


```python
#Add new column "OUTDEGREE" which is number of directly reachable airports i to j
# in case I run this cell more than once.
if 'OUTDEGREE' in segments.columns:
    del segments['OUTDEGREE']
    
segments['OUTDEGREE'] = segments.groupby(['ORIGIN_INDEX'])['ORIGIN_INDEX'].transform('count')
segments.head(10)
```

Now, Let's create a weighted column.  Assume that a random flyer is equally as likely (I know... not REALITY) to pick any of the destinations j directly reachable from i.  The weight is this random probability. 

Pij = 1/di

```python
segments['WEIGHT'] = 1.0 / segments['OUTDEGREE']
#SANITY CHECK.   These should sum to 1! 
segments.head(10)
```

Now, construct a sparse matrix P, corresponding to the state-transition matrix P.  The dimension of the matrix should be, n_airports by n_airports?


```python
import matplotlib.pyplot as plt
%matplotlib inline

def spy(A, figsize=(6, 6), markersize=0.5):
    """Visualizes a sparse matrix."""
    fig = plt.figure(figsize=figsize)
    plt.spy(A, markersize=markersize)
    plt.show()
```

```python
P = sp.sparse.coo_matrix((segments['WEIGHT'], \
                         (segments['ORIGIN_INDEX'], segments['DEST_INDEX'])),\
                        shape = (n_airports, n_airports))

spy(P)
```

At time  t=0t=0 , suppose the random flyer is equally likely to be at any airport with an outbound segment, i.e., the flyer is at one of the "actual" origins. Create a NumPy vector x0[:] such that x0[i] equals this initial probability of being at airport i.

Note: If some airport  ii  has no outbound flights, then be sure that  xi(0)=0xi(0)=0 .

```python
print("Number of actual locations (whether origin or destination):", n_actual)
```

```python
print("Number of actual airports in matrix:", n_airports)
```

```python
#Create empty vector (all zero's the length of # airports)
x0 = np.zeros(n_airports)
# print(x0)
#remember, the segments df has multiple entries for every origin and dest
actual_airport_origins = segments['ORIGIN_INDEX'].unique()
#change ech entry of x0 to be probability of being at airport i => 1.0 / n_actual
x0[actual_airport_origins] = 1.0 / n_actual


# Visually inspect your result:
def display_vec_sparsely(x, name='x'):
    i_nz = np.argwhere(x).flatten()
    df_x_nz = pd.DataFrame({'i': i_nz, '{}[i] (non-zero only)'.format(name): x[i_nz]})
    display(df_x_nz.head())
    print("...")
    display(df_x_nz.tail())
    
display_vec_sparsely(x0, name='x0')
```

```python
#What are the top 25 airports using Markov Analysis?

def eval_markov_chain(P, x0, t_max):
    ###
    x = x0
    for _ in range(t_max):
        x = P.transpose() * x
    return x
    
    ###
    
T_MAX = 50
x = eval_markov_chain(P, x0, T_MAX)
display_vec_sparsely(x)


print("\n=== Top 25 airports ===\n")
ranks = np.argsort(-x)
top10 = pd.DataFrame({'Rank': np.arange(1, 26),
                      'Code': airports.iloc[ranks[:25]]['IATA_CODE'],
                      'Description': airports.iloc[ranks[:25]]['AIRPORT'],
                      'x(t)': x[ranks[:25]],
                     'Latitude': airports.iloc[ranks[:25]]['LATITUDE'],
                     'Longitude': airports.iloc[ranks[:25]]['LONGITUDE'] })
top10[['x(t)', 'Rank', 'Code', 'Description', 'Latitude', 'Longitude']]
```

```python
#What are all of the airport rankings using Markov Analysis?

def eval_markov_chain(P, x0, t_max):
    ###
    x = x0
    for _ in range(t_max):
        x = P.transpose() * x
    return x
    
    ###
    
T_MAX = 50
x = eval_markov_chain(P, x0, T_MAX)
display_vec_sparsely(x)


print("\n=== Airport Rankings ===\n")
ranks = np.argsort(-x)
all_ranks = pd.DataFrame({'Rank': np.arange(1,len(ranks)+1),
                      'Code': airports.iloc[ranks]['IATA_CODE'],
                      'Description': airports.iloc[ranks]['AIRPORT'],
                      'x(t)': x[ranks],
                     'Latitude': airports.iloc[ranks]['LATITUDE'],
                     'Longitude': airports.iloc[ranks]['LONGITUDE'] })
all_ranks[['x(t)', 'Rank', 'Code', 'Description', 'Latitude', 'Longitude']]
```

```python
#Visualize the Airport Ranking from the Markov Analysis
#Resource:  https://www.geodose.com/2019/01/realtime-flight-tracking-pandas-bokeh-python.html
# DEFINE COORDINATE RANGE IN WEB MERCATOR COORDINATE SYSTEM (EPSG:3857)

#This is US range
x_range,y_range=([-15187814,-6458032], [2505715,6567666])

#To plot points... 
#Step 1:  convert the pandas df into bokeh column data source
#Step 2: Add the points using bokeh figure circle method

#FUNCTION TO CONVERT GCS WGS84 TO WEB MERCATOR
#https://www.geodose.com/2019/01/realtime-flight-tracking-pandas-bokeh-python.html
def wgs84_to_web_mercator(df, lon="Longitude", lat="Latitude"):
    k = 6378137
    df["x"] = df[lon] * (k * np.pi/180.0)
    df["y"] = np.log(np.tan((90 + df[lat]) * np.pi/360.0)) * k
    return df
wgs84_to_web_mercator(all_ranks)

#add scaler column for ranks
all_ranks['scaled_rank'] = all_ranks['x(t)'] * 1400

all_ranks.head()
```

```python
#SET COLOR PALETTE
color_mapper = linear_cmap(field_name='x(t)', palette = cividis(10) ,low=min(all_ranks['x(t)']),\
                                    high=max(all_ranks['x(t)']))

#BOKEH COLUMN DATA SOURCE
flight_source=ColumnDataSource(all_ranks)

#SETUP FIGURE
p=figure(x_range=x_range,y_range=y_range,x_axis_type='mercator',y_axis_type='mercator',
  sizing_mode='scale_width',plot_height=300,title = "Airports Ranks by Markov Process")
p.add_tile(tile_provider)

#Add Airports
p.circle('x','y',line_color = color_mapper, color = color_mapper, source=flight_source,\
         size= 'scaled_rank' ,fill_alpha=1,line_width=0.5)

#SETUP HOVER TOOL
my_hover = HoverTool()
my_hover.tooltips = [('Airport Code', '@Code'),('Description', '@Description'),('Rank', '@Rank')]
p.add_tools(my_hover)

#Setup Color Bar
color_bar = ColorBar(color_mapper = color_mapper['transform'], width = 25, location = (0,0))
p.add_layout(color_bar, 'right')
               
#SHOW FIGURE INLINE
show(p)

#OPEN FIGURE IN BROWSER
p
```
